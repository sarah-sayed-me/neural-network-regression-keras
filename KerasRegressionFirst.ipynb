```python
import pandas as pd 
import numpy as np 
import keras 
import warnings 
warnings.simplefilter('ignore',FutureWarning)
```

```python
filepath='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv'
data = pd.read_csv(filepath)
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
data_cols = data.columns
```

```python
predictors = data[data_cols[data_cols!='Strength']]
target = data['Strength']
```

```python
predictors.head()
```

```python
target.head()
```

```python
predictors_normalized = (predictors-predictors.mean())/predictors.std()
predictors_normalized.head()
```

```python
predictors_normalized.shape
```

```python
n_cols = predictors_normalized.shape[1]
```

```python
from keras.models import Sequential
from keras.layers import Dense 
from keras.layers import Input
```

```python
def regression_model():
    model = Sequential()
    model.add(Input(shape=(n_cols,)))
    model.add(Dense(50,activation='relu'))
    model.add(Dense(50,activation='relu'))
    model.add(Dense(1))

    model.compile(optimizer='adam',loss='mean_squared_error')
    return model 
```

```python
model = regression_model()
```

```python
model.fit(predictors_normalized,target,validation_split=0.3,epochs=100,verbose=2)
```
